## Q13. Train-Test Split (Manual)

**Task:** Write a function `train_test_split_manual(X, y, test_size=0.2, seed=42)`.

- Shuffle consistently
- Return X_train, X_test, y_train, y_test


````python
# Q13 - Train-Test Split (Manual)
# TODO: Write your solution below.

import random

def train_test_split_manual(X, y, test_size=0.2, seed=42):

  random.seed(seed)

  #combine x and y to keep features and lable together
  data = list(zip(X,y))

  #shuffle for random distribution
  random.shuffle(data)

  #now seperate X and y again
  X_shuffled, y_shuffled = zip(*data)

  #split point
  splitNum = int (len(X)- (1-test_size))

  #split data

  X_train = X_shuffled[:splitNum]
  X_test = X_shuffled[splitNum:]

  y_train = y_shuffled[:splitNum]
  y_test = y_shuffled[splitNum:]

  return X_train, X_test, y_train, y_test

````

## Q17. Precision & Recall (Manual)

**Task:** For binary classification, compute precision and recall.

Use:
`y_true=[1,0,1,1,0]`
`y_pred=[1,0,0,1,1]`


````python
# Q17 - Precision & Recall (Manual)
# TODO: Write your solution below.

y_true=[1,0,1,1,0]
y_pred=[1,0,0,1,1]

TP=FP=FN= 0

for i,j in zip(y_true, y_pred):
  if i==1 and j==1:
    TP += 1
  elif i==0 and j==1:
    FP += 1
  elif i==1 and j==0:
    FN += 1
precision = TP/(TP+FP)
recall = TP/(FN+TP)

print("Precision:", precision)
print("Recall:", recall)

# TODO
````

## Q25. Batch Prediction

**Task:** Modify prediction logic to accept multiple records and return list of predictions.


````python
# Q25 - Batch Prediction
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel, Field
import numpy as np
import pickle

# Create FastAPI app
app = FastAPI(title="Iris Batch Prediction API")

# Load trained model once at startup
with open("iris_model.pkl", "rb") as file:
    model = pickle.load(file)


# Request for batch prediction
class BatchPredictRequest(BaseModel):
    # Accept multiple records, Each inner list must contain 4 feature values
    features: list[list[float]] = Field(
        ...,
        description="List of records, each record must have 4 features"
    )


@app.get("/")
def home():
    return {"message": "Batch Prediction API is running"}


@app.post("/predict")
def predict(request: BatchPredictRequest):
    try:
        # Convert input list into NumPy array
        input_array = np.array(request.features)

        # input should be 2D
        if input_array.ndim != 2 or input_array.shape[1] != 4:
            raise HTTPException(
                status_code=400,
                detail="Each record must contain exactly 4 features"
            )

        # Predict for all input records
        predictions = model.predict(input_array)

        # Convert NumPy array to normal Python list
        return {"predictions": predictions.tolist()}

    except Exception as e:
        raise HTTPException(status_code=400, detail=str(e))

````

## Q27. Simple Cache

**Task:** Implement a cache dict that stores predictions for repeated inputs.


````python
# Q27 - Simple Cache
# TODO: Write your solution below.

from fastapi import FastAPI
from pydantic import BaseModel
import numpy as np
import pickle

app = FastAPI()

#Load Model
with open("iris_model.pkl", "rb") as f:
  model = pickle.load(f)

#cache dictionary
cache = {}

class PredictRequest(BaseModel):
  features: list[float]

@app.post("/predict")
def predict(request: PredictRequest):

  #convert input to tuple
  input_tuple = tuple(request.features)

  #check if prediction result already exist for the tuple
  if input_tuple in cache:
    return {
        "prediction": cache[input_tuple],
        "source": "cache"
    }

  #if not in cache
  input_array = np.array(request.features).reshape(1, -1)
  prediction = int(model.predict(input_array)[0])

  # Store in cache
  cache[input_tuple] = prediction

  return {
        "prediction": prediction,
        "source": "model"
    }
````

## Q28. Data Drift Check (mean shift)

**Task:** Given baseline feature means and new batch means, flag drift if abs diff > threshold.


````python
# Q28 - Data Drift Check (mean shift)
# TODO: Write your solution below.


import numpy as np

def check_drift(baseline_data, new_data, threshold=0.5):

  baseline_mean = np.mean(baseline_data)
  new_mean = np.mean(new_data)

  diff = abs(baseline_mean - new_mean)

  drift = diff > threshold

  return {
        "baseline_mean": baseline_mean,
        "new_mean": new_mean,
        "difference": diff,
        "drift_detected": drift
    }


````
